```python
import pandas
from pandas import DataFrame as df
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
data = pandas.read_csv('/Users/kuljeetsinghshekhawat/Downloads/cost_revenue_clean.csv')
data.describe()
```

```python
X = df(data,columns=['production_budget_usd'])
y = df(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X,y,alpha=0.3)
plt.title('Film cost vs Global revenue')
plt.xlabel('production budget')
plt.ylabel('worldwide collection')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

```python
regression = LinearRegression()
regression.fit(X,y)
```

slope coefficient

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X,y,alpha=0.3)
plt.plot(X,regression.predict(X), color = 'red', linewidth =2)
plt.title('Film cost vs Global revenue')
plt.xlabel('production budget')
plt.ylabel('worldwide collection')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)

plt.show()
```

```python
regression.score(X,y)
```
